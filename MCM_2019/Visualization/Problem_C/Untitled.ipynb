```python
import pandas  as pd
import numpy as np
import pyecharts as pc
from matplotlib import pyplot as plt 
```

```python
Distance = pd.read_csv('distanceDataFrame.csv')
Profile = pd.read_csv('Drug_Use_Profile_in_2011.csv')
```

```python
def Listify_df(df, col_name):
    res = []
    for i in range(0,df.shape[0]):
        res.append(df[col_name][i])
    return res
```

```python
def Count_Neignbor_County(FIPS, distance):
    L = Listify_df(Distance, FIPS)
    count = 0
    for i in range(0, len(L)):
        if(L[i]<= distance):
            count = count + 1
    return count
```

```python
def min_index(List):
    index = 0
    if List[index]==0:
        index = index + 1
    
    for i in range(0, len(List)):
        if List[i]<=List[index] and List[i]>0:
            index = i
    return index
```

```python
def Find_Nearest_County(FIPS):
    L = Listify_df(Distance, FIPS)
    min_in = min_index(L)
    return Distance["%s"%FIPS][min_in]
```

```python
County_list = Distance.columns
S = 0
for county in County_list:
    S += Find_Nearest_County(county)
S/len(County_list)
```

# 添加邻近郡的数量：

```python
#添加邻近郡的数量：
Neighber_count = []
for i in range(0, Profile.shape[0]):
    FIPS = Profile['FIPS_com'][i]
    
    count = Count_Neignbor_County("%d"%FIPS, 50000)
    Neighber_count.append(count)
```

```python
Neighber_Count = pd.DataFrame(Neighber_count)
```

```python
Neighber_Count.columns = ['Neighber_County_Count']
```

```python
Profile_c = pd.merge(Profile,Neighber_Count, left_index=True,right_index=True,how='outer')
```

```python
Profile_c
```

# 添加最邻近郡：

```python
#添加邻近郡的数量：
Nearest_county = []
for i in range(0, Profile.shape[0]):
    FIPS = Profile['FIPS_com'][i]
    
    c = Find_Nearest_County('%d'%FIPS)
    Nearest_county.append(c)
```

```python
Nearest_County = pd.DataFrame(Nearest_county)
```

```python
Nearest_County
```

```python
Nearest_County.columns = ['Nearest_County']
```

```python
Profile_b = pd.merge(Profile_c,Nearest_County, left_index=True,right_index=True,how='outer')
```

# 添加最邻近的城市的Drug_Use

```python
def find_County_Drug_Use(FIPS):
    for i in range(0, Profile.shape[0]):
        if Profile['FIPS_com'][i] == FIPS:
            return Profile['Drug_Use'][i]
```

```python
#添加NC 的 Drug_Use：
NC_Drug_Use = []
for i in range(0, Profile_b.shape[0]):
    NC_Drug_Use.append(find_County_Drug_Use(Profile_b['Nearest_County'][i]) )
```

```python
NC_Drug_Use = pd.DataFrame(NC_Drug_Use)
```

```python
NC_Drug_Use.columns = ['NC_Drug_Use']
```

```python
Profile_a = pd.merge(Profile_b,NC_Drug_Use, left_index=True,right_index=True,how='outer')
```

# 添加最近城市的Report Time

```python
def find_County_RT(FIPS):
    for i in range(0, Profile.shape[0]):
        if Profile['FIPS_com'][i] == FIPS:
            return Profile['Reports_Time'][i]
```

```python
#添加NC 的 Drug_Use：
NCRT = []
for i in range(0, Profile_b.shape[0]):
    NCRT.append(find_County_RT(Profile_b['Nearest_County'][i]) )
```
