```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import minmax_scale
```

```python
name='2010'
ct=pd.read_csv('%s.csv'%name)
ct_np=np.array(ct)[1:,1:]
```

```python
ct_np_sc = minmax_scale(ct_np)
```

```python
ct_np_sc
```

```python
class Cf():
    def __init__(self,lbd=0.01,itr=100,cls=1,alpha=0.01): 
        self.lbd=lbd
        self.itr=itr 
        self.cls=cls 
        self.alpha=alpha 
        self.cost=np.zeros(itr) 
        self.year=[] 
        self.feature=[]
        
    def fit(self,x): 
        self.year=0.01*np.random.rand(x.shape[1],self.cls) 
        print(self.year)
        self.feature= 0.01*np.random.rand(x.shape[0],self.cls)
        print(self.feature)
        
        for i in range(self.itr): 
            self.cost[i]=self.cost_fun(self.year,self.feature,x) 
            print(self.cost[i])
            print(i)
            for ny in range(self.year.shape[0]-1):
                for nf in range(self.feature.shape[0]-1):
#                     print(ny, nf)
                    self.year[ny,:] -= self.alpha * (np.dot(self.feature[nf,:].transpose(),self.year[ny,:]) - x[nf,ny]) * self.year[ny,:]
            for nf in range(self.feature.shape[0]-1):
                for ny in range(self.year.shape[0]-1):
#                     print(ny, nf)
                    self.feature[nf,:]-=self.alpha * np.dot( np.dot(self.feature[nf,:].transpose(),self.year[ny,:]) - x[nf,ny] , self.feature[nf,:])
        
        
                                                                 
    def cost_fun(self,year,feature,x):
        y = np.dot(feature,year.transpose())
#         print(y)
#         print(x)
#         print(y-x)
        Cost  = (y-x)**2
        cost=1/2*sum(sum(Cost))
#         print(Cost)
#         print(cost)
        return cost
```
