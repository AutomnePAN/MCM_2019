```python
def find_Potentiel_city(year, l):
    res = []
    
    for i in range(Potentiel[year - 2011].shape[0]):
        if(Potentiel[year - 2011]['is_Source_Rank'][i] == l):
            res.append(find_County_from_FIPS(Potentiel[year - 2011]['FIPS'][i]))
    
    res = pd.DataFrame(res, columns = ['County', 'State'])
    return res
```

```python
def find_Drug_Using_city(year,l):
    res = []
    for i in range(Potentiel[year - 2011].shape[0]):
            if(Potentiel[year - 2011]['Drug_Use_Rank'][i] == l):
                res.append(find_County_from_FIPS(Potentiel[year - 2011]['FIPS'][i]))
            
    res = pd.DataFrame(res, columns=['County', 'State'])
    return res  
```
