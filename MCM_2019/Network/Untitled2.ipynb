```python
def find_Proportion(P):
    Res = [0,0,0,0]
    for i in range(0, P.shape[0]):
        for k in range(4):
            if P['Drug_Use_Rank'][i] == k:
                Res[k] = Res[k]+1
    S = sum(Res)
    for k in range(4):
        Res[k] = Res[k]/S
    return Res
```

```python
def Listify_df(df, col_name):
    res = []
    for i in range(df.shape[0]):
        res.append(df[col_name][i])
    return res
```

```python
def find_County_from_FIPS(FIPS):
    res = []
    for i in range(0, County.shape[0]):
        if County['FIPS_com'][i] == FIPS:
            res.append(County['County'][i])
            res.append(County['State'][i])
            return res
```

```python
def is_in_list(List, num):
    flag = 0 
    for i in range(0, len(List)):
        if List[i] == num:
            flag  = 1
    return flag
```

```python
def min_index_n(List, n):
   
    res = []
    size = 0
    
    for k in range(0, n):
        index = 0
        
        while(is_in_list(res, index) == 1):
            index = index + 1
        
        for i in range(0, len(List)):
         
    #           print("å½“å‰indexä»¥åŠi")
    #             print(List[index])
    #             print(List[i])
            
            if List[i] <= List[index] :
#                 print("List i å°äºŽç­‰äºŽäºŽ List_index")
                
                if (is_in_list(res, i) == 0):
                    index = i
                
        res.append(index)
        size = size +1
    return res
```

```python
def Find_Nearest_n_County(FIPS, n):
    L = Listify_df(Distance, "%d"%FIPS)
    CL = min_index_n(L,n)
    res = []
    for i in range(len(CL)):
        res.append(Distance["FIPS"][CL[i]])
    return res
```

```python
def find_County_index(FIPS):
    L = Listify_df(Profile, 'FIPS')
    for i in range(0, len(L)):
        if L[i] == FIPS:
            return i
```
